```python
#alternative way to create Counter for frequency_histogram

#importing nessessary components
import random as rnd

#Create population
p=[rnd.randint(0,100) for i in range(100)]
#Creating counter array which is an analogy of Counter object
#It consists from 1d arrays which representing number and count of this number
print(p)
c=[]
def if_number_in_Counter(i,c):#function that returns "No" if number not in counter and element id  where number is if number in counter
    for j in range(len(c)):
        if c[j][0]==i:
            return j   
    return "No"       
for i in p:
    if if_number_in_Counter(i,c)!="No":
        c[if_number_in_Counter(i,c)][1]+=1 
    else:
        c.append([i,1])
        
unique_values=[] #Creating empty array of unique values
count_of_values=[]#Creating empty array of count of values
for i in p:#creating array of unique values
    if (i in unique_values)==0:#if current element of population not in unique_values, appending current element to our array of unique values
        unique_values.append(i)
for j in unique_values:#creating array of count of values
    count=0
    for i in p:
        if i==j:
            count+=1
    count_of_values.append(count)
    
print(c)
```

```python
#creating population with a specific mean value and specific number of objects
import random as rnd
import statistics as st
p=[]
rem_sum=2000.0;
k=100
#We want to create population with mean equals 20.0
first=90#number of elements which generated by first pattern of generation pmethod
second=k-first#number of elements which generated by second pattern of generation method
for i in range(first):
    el=rnd.randint(0,rem_sum//first)
    p.append(el)
    rem_sum-=el
for i in range(second):
    el=rem_sum/second
    p.append(el)

print(p)
print(st.mean(p))
```

```python
from matplotlib import pyplot as plt
import statistics as st
plt.plot([4,5,2,3,1],[2,1,4,3,5])
plt.show()
def cor(x,y):
    n=len(x)
    sum_products=0
    mean_x=st.mean(x)
    mean_y=st.mean(y)
    sd_x=st.stdev(x)
    sd_y=st.stdev(y)
    for i in range(len(x)):
        pr_of_dif=(x[i]-mean_x)*(y[i]-mean_y)
        sum_products+=pr_of_dif
    print(sum_products)
    r_xy=sum_products/(sd_x*sd_y*(n-1))
    return r_xy
print(round(cor([4,5,2,3,1],[2,1,4,3,5])))
```
